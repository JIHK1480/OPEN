```python
import pandas as pd
```

```python
data = {
    '이름' : ['A', 'B', 'C'],
    '수학' : [40, 60, 70]
}

data2 = {
    '이름' : ['A', 'C'],
    '과학' : [50, 60],
    '언어' : [80, 90]
}

data = pd.DataFrame(data)
data2 = pd.DataFrame(data2)
```

```python
data
```

```python
data2
```

```python
merge = pd.merge(data, data2, how="left", on="이름")
merge
```

```python
merge = pd.merge(data, data2, how="right", on="이름")
merge
```
